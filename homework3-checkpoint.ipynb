```python
# bài 9:nhập vào số nguyên dương n, kiểm tra xem a có phải số chính phương hay không? số chính phương là số có căn bậc 2 là số nguyê
n = int(input(" nhập số nguyên dương "))
def is_sqrt(n):
    if n<0:
        return False
    x =round( n ** (0.5))
    return x *x ==n
print( is_sqrt(n))
    



```

```python
# bài 14:nhập dãy số nguyên . dừng nhập khi có -1.sau khi nhập xong in số max, min
nums = []
while True:
    n = int(input())
    if n == -1:
        break
        nums.append(n)
print( " số lớn nhất " , max ( nums))
print(" số nhỏ nhất ", min( nums))
```

```python
# bài 16: nhập vào số nguyên dương. đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ
n = int(input(" nhập vào số nguyên dương "))
chu_so_le = 0
chu_so_chan = 0
while n>0:
    d = n% 10
    n //=10
    if d %2==0:
        chu_so_chan +=1
    elif d %2 ==1:
        chu_so_le +=1
print( " số chữ số chẵn ", chu_so_chan )
print( " số chữ số lẻ ", chu_so_le )
```
