```python
def Max_Of_Two(a:int,b:int):
    if a>b:
        return a
    else:
        return b
print(Max_Of_Two(2.5,8))
```

```python
def is_prime(n):
    if n<2:
        return False
    for i in range( 2, int(n**0.5)):
        if n%i==0:
            return False
    return True
print(is_prime(5))
        
        
```

```python
def is_perfect(n):
    tong=0
    for i in range(1, n+1):
        if n%i==0:
            tong+=i
            if n == tong:
                return True
    return False
print(is_perfect(10)) 
            
```

```python
k =int(input())
v = list(map(int, input().split()))

def vi_tri( v,k):
    for i in range(len(v)):
        if v[i]==k:
            return i
    return -1 
print(vi_tri(v,k))
```

```python
def giai_thua(n):
    x= 1
    for i in range(1, n+1):
        x *=i
    return x
print(giai_thua(6))
```
