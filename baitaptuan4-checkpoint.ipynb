```python
# nhập vào số nguyên dương. lặp lại phép n = n + reserve đến khi n là số đối xứng. in n đối xứng và số lần cộng
n = int(input("nhập số nguyên dương "))
def reserve(n):
    r_sv=0
    while n>0:
        r_sv= r_sv*10+ n%10
        n = n//10
    return r_sv
d = 0
while n!= reserve(n):
    n = n + reserve(n)
    d +=1
print(d, reserve(n))

```

```python
#w4a9: nhập vào số nguyên dương, in tất cả số chính phương<=n đôi một khác nhau
from math import sqrt
n = int(input())
def check(x):
    cnt = [0]*10
    while x:              # lặp lại không ngừng cho đến khi x =0
        t = x %10
        cnt[t] +=1
        x //=10
        if cnt[t]>1:
            return False
    return True
for i in range ( 1,int( sqrt(n) +1)):
    if check(i*i):
        print( i*i, end =' ')

        
        
```

```python
#w4a10
n = int(input())
ans =0                                   # chức năng: so sánh + lưu số lần đếm max
val=0                                    # lưu giá trị x max trong lệnh if so sánh 
for x in range (1,n +1):
    res=0
    original=x                          # lưu cố định giá trị của x trong mỗi vòng lặp for, không thì while sẽ rất ảnh hưởng
    while x !=1:
        if x %2==0:
            x=x/2
        elif x %2==1:
            x = x*3+1
        res +=1
    if res > ans:                      # không phải x càng lớn thì res càng lớn, res tùy, so sánh với 1 gtri nào đó nếu lớn hơn thì val chính = số x đã cố định
        ans = res
        val = original
print(val, ans)
        
```
